```python
# Vamos utilizar um modelo j treinado para reconhecer faces. Primeiro, baixe o arquivo xml que est no Github: https://github.com/opencv/opencv/tree/master/data/haarcascades
# Encontre o haarcascade_frontalface_alt.xml. Aps clicar nele, clique em Raw, depois clique com o boto direito do mouse salvar como...

import cv2
import numpy as np
from matplotlib import pyplot as plt

imagem = cv2.imread('C:/Users/Natanael/Documents/DidaticaTech/pessoas.jpg')
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

classificador = cv2.CascadeClassifier('C:/Users/Natanael/Documents/DidaticaTech/haarcascade_frontalface_alt.xml')
retangulos_rostos = classificador.detectMultiScale(imagem_cinza, scaleFactor=1.1, minNeighbors=4) # vai retornar as dimenses dos retngulos das faces
# scaleFactor define a proporo de redimensionamento da imagem na gerao de pirmides de imagens. Se colocarmos 1.1, significa
# que o algoritmo ir redimensionar a imagem por uma proporo de 10% para cima ou para baixo vrias vezes at atingir um limite
# de dimenses X, Y. Depois, ele ir tentar detectar faces em todas essas imagens de diferentes escalas geradas. Se colocarmos
# 1.05, a pirmide de imagens ser gerada de 5% em 5%, e assim por diante. 

# minNeighbors especifica quantos vizinhos cada retngulo candidato deve ter para definir a deteco de uma face. Quanto maior 
# esse valor, menos deteces se ter, mas mais garantido que a deteco ser uma face. De 3-6 costumam ser bons valores para 
# se ajustar.  

for (x,y,largura,altura) in retangulos_rostos:
    imagem = cv2.rectangle(imagem,(x,y), (x+largura,y+altura), (255,0,0), 4)
    
plt.figure(figsize=(10,10))
plt.imshow(imagem)
plt.show()
```
